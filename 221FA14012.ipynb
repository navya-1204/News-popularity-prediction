```python
# prompt: # prompt: from the above data in the file give me linear regression between TS21	TS22	in python and also represent the title in that graph

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
df = pd.read_csv("/content/LinkedIn_Obama.csv")

# Ensure the 'TS21' and 'TS22' columns exist and are numeric
if 'TS21' in df.columns and 'TS22' in df.columns:
    x = pd.to_numeric(df['TS21'], errors='coerce')
    y = pd.to_numeric(df['TS22'], errors='coerce')

    # Remove rows with NaN values in 'TS21' or 'TS22'
    valid_data = df.dropna(subset=['TS21', 'TS22'])
    x = valid_data['TS21']
    y = valid_data['TS22']

    # Check if there are enough valid data points for regression
    if len(x) > 1 and len(y) > 1:
        # Calculate the linear regression line
        coefficients = np.polyfit(x, y, 1)  # Degree 1 polynomial for linear regression
        polynomial = np.poly1d(coefficients)  # Create the polynomial function
        regression_line = polynomial(x)  # Generate y-values based on the regression line

        # Create the plot
        plt.figure(figsize=(10, 6))  # Increase the figure size for better readability
        plt.scatter(x, y, color='blue', label='Data Points', alpha=0.7)  # Scatter plot with data points
        plt.plot(x, regression_line, color='red', label='Linear Regression')  # Regression line

        # Add labels and title
        plt.xlabel('TS21', fontsize=12)
        plt.ylabel('TS22', fontsize=12)
        plt.title('Linear Regression between TS21 and TS22', fontsize=14)
        plt.legend()

        # Display the plot
        plt.grid(True)  # Add gridlines for better readability
        plt.tight_layout()  # Adjust layout to fit elements
        plt.show()
    else:
        print("Not enough valid data points for regression.")
else:
    print("The specified columns 'TS21' and 'TS22' are not in the DataFrame.")


```

```python
# prompt: # prompt: from the above data in the file give me linear regression between SentimentTitle	SentimentHeadline	in python and also represent the title in that graph

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
df = pd.read_csv("/content/News_Final.csv")

# Ensure the 'SentimentTitle' and 'SentimentHeadline' columns exist and are numeric
if 'SentimentTitle' in df.columns and 'SentimentHeadline' in df.columns:
    # Convert columns to numeric (in case there are invalid values)
    x = pd.to_numeric(df['SentimentTitle'], errors='coerce')
    y = pd.to_numeric(df['SentimentHeadline'], errors='coerce')
    title = df['Title']  # Get the 'Title' column for labeling

    # Remove rows with NaN values in 'SentimentTitle' or 'SentimentHeadline'
    valid_data = df.dropna(subset=['SentimentTitle', 'SentimentHeadline'])
    x = valid_data['SentimentTitle']
    y = valid_data['SentimentHeadline']
    title = valid_data['Title']

    # Check if there are enough valid data points for regression
    if len(x) > 1 and len(y) > 1:
        # Calculate the linear regression line
        coefficients = np.polyfit(x, y, 1)  # Degree 1 polynomial for linear regression
        polynomial = np.poly1d(coefficients)  # Create the polynomial function
        regression_line = polynomial(x)  # Generate y-values based on the regression line

        # Create the plot
        plt.figure(figsize=(10, 6))  # Increase the figure size for better readability
        plt.scatter(x, y, color='blue', label='Data Points', alpha=0.7)  # Scatter plot with data points
        plt.plot(x, regression_line, color='red', label='Linear Regression')  # Regression line

        # Add labels and title
        plt.xlabel('SentimentTitle', fontsize=12)
        plt.ylabel('SentimentHeadline', fontsize=12)
        plt.title('Linear Regression between SentimentTitle and SentimentHeadline', fontsize=14)

        # Add a conditional annotation for a subset of points to avoid clutter
        for i in range(0, len(x), max(1, len(x)//20)):  # Annotate roughly 20 points
            plt.annotate(title.iloc[i], (x.iloc[i], y.iloc[i]), fontsize=9, alpha=0.75)

        plt.legend()
        plt.grid(True)
        plt.tight_layout()
        plt.show()
    else:
        print("Not enough valid data points for regression.")
else:
    print("The specified columns 'SentimentTitle' and 'SentimentHeadline' are not in the DataFrame.")
```
